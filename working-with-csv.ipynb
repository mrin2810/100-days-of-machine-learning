## Opening a csv file with URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/thephpleague/csv/master/test_files/prenoms.csv"
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11'}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

pd.read_csv(data, sep=";")
```

## Encoding

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day15%20-%20working%20with%20csv%20files/zomato.csv"
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11'}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

pd.read_csv(data, encoding='latin-1')
```
