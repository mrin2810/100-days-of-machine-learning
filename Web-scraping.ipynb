```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 14_1) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.1 Safari/605.1.15'}

# We pass these headers as headers in order to show that they are coming from a browsert
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
comapanies = soup.find_all('div', class_='companyCardWrapper')
```

```python
ratings = []
name = []
no_of_reviews = []
for company in comapanies:
    name.append(company.find('h2', class_='companyCardWrapper__companyName').text.strip())
    ratings.append(company.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
    no_of_reviews.append(company.find('div', class_='companyCardWrapper__tertiaryInformation')
                         .find('a', class_='companyCardWrapper__ActionWrapper')
                         .find('span', class_='companyCardWrapper__ActionCount').text.strip())
```

```python
for i in range(len(name)):
    print(name[i], ratings[i], no_of_reviews[i])
```

```python
import pandas as pd
d = {
    'name': name,
    'ratings': ratings,
    'review_count': no_of_reviews
}
df = pd.DataFrame(d)
```
