```python
import folium
import pandas as pd
import matplotlib.pyplot as plt

# Sample crash data (replace this with your actual data)
data_2019 = {
    'Date': pd.date_range('2019-06-01', '2019-08-31', freq='D'),
    'Latitude': [40.7128] * 92,  # Sample latitude data for 2019
    'Longitude': [-74.0060] * 92,  # Sample longitude data for 2019
}

data_2020 = {
    'Date': pd.date_range('2020-06-01', '2020-08-31', freq='D'),
    'Latitude': [40.7128] * 92,  # Sample latitude data for 2020
    'Longitude': [-74.0060] * 92,  # Sample longitude data for 2020
}

df_2019 = pd.DataFrame(data_2019)
df_2020 = pd.DataFrame(data_2020)


# Plot time series
plt.figure(figsize=(12, 6))
plt.plot(df_2019.groupby(by='Date').count()['Latitude'], label='2019', marker='o')
plt.plot(df_2020['Date'], label='2020', marker='o')
plt.title('Crash Frequency Over Summer (2019 vs. 2020)')
plt.xlabel('Date')
plt.ylabel('Crash Count')
plt.legend()
plt.show()

# Create Folium map
map_center = [df_2019['Latitude'].mean(), df_2019['Longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=10)

# Add markers for 2019
for index, row in df_2019.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f'2019: {row["Date"].strftime("%Y-%m-%d")}').add_to(my_map)

# Add markers for 2020 with a different color
for index, row in df_2020.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']],).add_to(my_map)

# Display the map
my_map
```
