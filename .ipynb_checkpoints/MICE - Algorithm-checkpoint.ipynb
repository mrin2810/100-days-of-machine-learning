```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('datasets/50_startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)

np.random.seed(9)

df = df.sample(5)
df = df.iloc[:, 0:-1]
df

missing_values = [1, 3, 4]
```

```python
"""
mice function here, will do iterations on the data to predict the data correctly, 
and return the number of iterations required.

data - dataframe
missing_values - list of lists
where i_th list will have values that are supposed to go missing
"""
def mice(data, missing_values):
    df = data.copy()
    # Introduce missing values
    for j, i in enumerate(missing_values):
        df.iloc[i, j] = np.NaN
    # Stage 1: Replace all the NaNs with mean of respective columns
    df.iloc[:,0].fillna(df.iloc[:, 0].mean(), inplace=True)
    df.iloc[:,1].fillna(df.iloc[:, 1].mean(), inplace=True)
    df.iloc[:,2].fillna(df.iloc[:, 2].mean(), inplace=True)

    # Create a copy of the current DF
    prev_df = df.copy()

    range_list = [i for i in range(5)]
    col_list = [i for i in range(len(df.columns))]
    print("Iteration 0")
    print(df)
    i = 1
    while i <= 50:
        if i != 1 and (((prev_df - df) == 0).all().all()):
            break
        prev_df = df.copy()
        for j in range(len(df.columns)):
            X = df.iloc[filter(lambda x: x != missing_values[j], range_list), filter(lambda x: x != j, col_list)]
            y = df.iloc[filter(lambda x: x != missing_values[j], range_list), j]
            lr = LinearRegression()
            lr.fit(X, y)
            df.iloc[missing_values[j], j] = round(lr.predict(df.iloc[missing_values[j], filter(lambda x: x != j, col_list)].values.reshape(1, 2))[0], 2)
        print("Iteration " + str(i))
        print(df)
        i += 1   
    return i
    
    
print(mice(df, missing_values))
```
